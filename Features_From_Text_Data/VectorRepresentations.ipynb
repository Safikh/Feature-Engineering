# CountVectorizer

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
train_text = ["A bird in the hand is worth two in the bush",
                "Good things come to those who wait",
                "These watches cost $1500!",
                "There are other fish in the sea",
                "The ball is in your court",
                "Mr. Smith Goes to Washington",
                "Doogie Howser M.D."]
count_vectorizer = CountVectorizer()
```

```python
count_vectorizer.fit(train_text)
```

```python
count_vectorizer.get_feature_names_out()
```

```python
count_vectorizer.get_stop_words()
```

```python
count_vectorizer.vocabulary_
```

```python
count_vectorizer.vocabulary_.get("things")
```

```python
text_vec = count_vectorizer.transform(train_text)
text_vec.toarray()
```

```python
test_text = ["Every cloud has a silver lining"] # None of these words are in the vocabulary
count_vectorizer.transform(test_text).toarray()
```

# NGrams CountVectorizer

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
bigram_vectorizer = CountVectorizer(ngram_range=(2,2)) # start and end of the n-value for n-grams. Only bigrams for (2,2)

transformed_vec = bigram_vectorizer.fit_transform(train_text + sentences)
```

```python
bigram_vectorizer.vocabulary_
```

```python
transformed_vec.toarray()
```

```python
n_gram_vec = CountVectorizer(ngram_range=(1,2)) # Both single words and bigrams
transformed_vec = n_gram_vec.fit_transform(train_text + sentences)
n_gram_vec.vocabulary_
```

```python
n_gram_vec = CountVectorizer(ngram_range=(2, 3)) # Both bigrams and trigrams
transformed_vec = n_gram_vec.fit_transform(sentences)
vocab = n_gram_vec.vocabulary_
vocab
```

```python
vocab.get("marie curie")
```

```python
word_count = transformed_vec.toarray().sum(axis=0)
word_count
```

```python
sorted_word_list = sorted([(word_count[i], n_gram) for n_gram, i in vocab.items()], reverse=True)
sorted_word_list
```

```python
from nltk import bigrams, trigrams
from nltk.tokenize import word_tokenize
```

```python
word_tokens = word_tokenize(" ".join(sentences))
word_tokens
```

```python
nltk_bigrams = bigrams(word_tokens)
list(nltk_bigrams)
```

```python
nltk_trigrams = trigrams(word_tokens)
list(nltk_trigrams)
```

```python
from nltk import ngrams
nltk_ngrams = ngrams(word_tokens, 2)
list(nltk_ngrams)
```

```python
from nltk.collocations import BigramCollocationFinder, BigramAssocMeasures
```

```python
bigram_measures = BigramAssocMeasures() # contains functions for calculating bigram association measures
finder = BigramCollocationFinder.from_words(word_tokens) # contains functions for finding bigrams 
finder.apply_freq_filter(3) # Bigrams with less than 3 occurances are removed
```

```python
matches = finder.nbest(bigram_measures.raw_freq, 15)
matches
```

# TF-IDF Vectorizer

```python
from sklearn.feature_extraction.text import TfidfTransformer
```

```python
tfidf_transformer = TfidfTransformer()
```

```python
count_vec = CountVectorizer()
transformed_vec = count_vec.fit_transform(sentences)
tfidf1 = tfidf_transformer.fit_transform(transformed_vec) # TfidfTransformer needs a sparse frequency matrix as input
```

```python
tfidf1.shape
```

```python
tfidf1.toarray()
```

```python
# Tokens can directly be passed to the TfidfVectorizer without needing a CountVectorizer
# TfidfVectorizer is a wrapper around CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
tfidfvec = TfidfVectorizer()
tfidf2 = tfidfvec.fit_transform(sentences)
tfidf2.shape
```

```python
tfidf2.toarray()
```

```python
tfidfvec.idf_ # inverse document frequency for all the words in the vocabulary
```

```python
dict(zip(tfidfvec.get_feature_names_out(), tfidfvec.idf_))
```
