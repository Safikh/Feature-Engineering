```python
import pandas as pd
import numpy as np

from sklearn import datasets

import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
```

```python
X, color = datasets._samples_generator.make_swiss_roll(n_samples=2000)

print(f'Shape of X: {X.shape}')
print(f'Shape of color: {color.shape}')
```

```python
X = pd.DataFrame(X)

ax = plt.subplots(figsize=(10, 10))
ax = plt.axes(projection='3d')

ax.scatter3D(X[0], X[1], X[2], c=color, cmap=plt.cm.Spectral)
```

```python
def apply_manifold_learning(X, method):

    X = method.fit_transform(X)

    print(f'New shape of X: {X.shape}')

    X = pd.DataFrame(X)

    plt.subplots(figsize=(10, 10))
    plt.axis('equal');

    plt.scatter(X[0], X[1], c=color, cmap=plt.cm.Spectral)
    plt.xlabel('X[0]')
    plt.ylabel('X[1]')

    return method
```

```python
from sklearn.manifold import MDS, LocallyLinearEmbedding
```

```python
mds = apply_manifold_learning(X, MDS(n_components=2, metric=False))
```

```python
mds = apply_manifold_learning(X, MDS(n_components=2, metric=True))
```

```python
lle = apply_manifold_learning(X, LocallyLinearEmbedding(n_components=2, n_neighbors=15, method='standard'))
```

```python
hlle = apply_manifold_learning(X, LocallyLinearEmbedding(n_components=2, n_neighbors=15, method='hessian'))
```
