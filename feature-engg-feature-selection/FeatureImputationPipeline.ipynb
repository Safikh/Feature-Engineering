```python
import numpy as np
import pandas as pd

from sklearn.impute import SimpleImputer
```

```python
diabetes = pd.read_csv('datasets/diabetes_processed.csv')
diabetes.head(10)
```

```python
features = diabetes.drop('Outcome', axis=1)
labels = diabetes['Outcome']
features.head(10)
```

```python
mask = np.random.randint(0, 100, size=features.shape).astype(np.bool)
mask = np.logical_not(mask)
```

```python
features[mask] = np.nan
features.sample(15)
```

```python
from sklearn.pipeline import make_pipeline
from sklearn.compose import ColumnTransformer

from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
```

```python
X_train, X_test, y_train, y_test = train_test_split(features, labels, test_size=0.2, random_state=0)
```

```python
transformer = ColumnTransformer(
    transformers=[('features', SimpleImputer(strategy='mean'), [0, 1, 2, 3, 4, 5, 6, 7])]
)
```

```python
clf = make_pipeline(transformer, DecisionTreeClassifier(max_depth=4))
clf = clf.fit(X_train, y_train)
clf.score(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score
y_pred = clf.predict(X_test)
accuracy_score(y_test, y_pred)
```
