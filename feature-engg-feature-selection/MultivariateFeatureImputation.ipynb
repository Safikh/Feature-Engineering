```python
df = pd.read_csv('datasets/diabetes_processed_incomplete.csv')
df.head(10)
```

```python
df.isnull().sum()
```

```python
features = df.drop(['Outcome'], axis=1)
label = df['Outcome']
features.head(10)
```

```python
imp = IterativeImputer(max_iter=10000, random_state=0)
imp.fit(features)
```

```python
features_arr = imp.transform(features)
features_arr.shape
```

```python
features = pd.DataFrame(features_arr, columns=features.columns)
features.head(10)
```

```python
features.isnull().sum()
```

```python
df = pd.concat([features, label], axis=1)
df.head(10)
```

```python
df.isnull().sum()
```
