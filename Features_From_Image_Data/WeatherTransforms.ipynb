```python
from skimage import io
import numpy as np
import matplotlib.pyplot as plt
import cv2
import albumentations as albu
```

```python
image = io.imread('data/images/taj_mahal_india.jpg')
plt.figure(figsize=(10, 10))
plt.imshow(image)
plt.show();
```

```python
def augment_and_show(aug, image):
    image_aug = aug(image=image)['image']
    plt.figure(figsize=(10, 10))
    plt.imshow(image_aug)
    plt.show();
```

```python
img_rain = albu.RandomRain(p=1, brightness_coefficient=0.9, drop_width=3, blur_value=1)
augment_and_show(img_rain, image)
```

```python
img_snow = albu.RandomSnow(p=1, brightness_coeff=1.9, snow_point_lower=0.3, snow_point_upper=0.5)
augment_and_show(img_snow, image)
```

```python
img_sunflare = albu.RandomSunFlare(p=1, flare_roi=(0, 0, 0.5, 0.5), angle_lower=0.6)
augment_and_show(img_sunflare, image)
```

```python
img_fog = albu.RandomFog(p=1, alpha_coef=0.1, fog_coef_lower=0.5, fog_coef_upper=0.8)
augment_and_show(img_fog, image)
```
