# Dictionary Learning

```python
import skimage
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits
from skimage.transform import resize
```

```python
digits = load_digits()
```

```python
digits.data.shape
```

```python
NUM_IMAGES = 48
```

```python
sample = digits.data[:NUM_IMAGES, :]
```

```python
fig = plt.figure(figsize=(12, 10))
for i in range(NUM_IMAGES):
    ax = fig.add_subplot(6, 8, i + 1, xticks=[], yticks=[])
    ax.imshow(sample[i].reshape((8, 8)), cmap='Blues_r',
                interpolation='nearest')
```

```python
from sklearn.decomposition import DictionaryLearning
dict_learn = DictionaryLearning(n_components=36,
                                    fit_algorithm='lars',
                                    transform_algorithm='lasso_lars')

X_dict = dict_learn.fit_transform(sample)
print(X_dict.shape)
```

```python
resized = resize(X_dict[0].reshape((6, 6)), (8, 8))
```

```python
plt.imshow(resized, cmap='Blues_r', interpolation='nearest')
```

```python
fig = plt.figure(figsize=(12, 10))
for i in range(NUM_IMAGES):
    ax = fig.add_subplot(6, 8, i + 1, xticks=[], yticks=[])
    ax.imshow(X_dict[i].reshape((6, 6)), cmap='Blues_r',
                interpolation='nearest')
```
