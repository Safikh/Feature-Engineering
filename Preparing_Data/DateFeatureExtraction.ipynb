```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
import datetime
```

```python
from pandas.plotting import register_matplotlib_converters
register_matplotlib_converters() # To visualize the date in the graph
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/matplotlib/sample_data/master/aapl.csv')
df.head()
```

```python
df = df.sample(frac=1, random_state=1) # Shuffle the data
df.reset_index(inplace=True, drop=True) # Reset the index
df.head()
```

```python
df.info()
```

```python
df['Date'] = pd.to_datetime(df['Date']) # Convert the date column to datetime
df.head()
```

```python
df.info()
```

```python
df.shape
```

```python
df.sort_values(by='Date', inplace=True) # Sort the data by date
df.head()
```

```python
df.reset_index(inplace=True, drop=True) # Reset the index
df.head()
```

```python
df['Date'] = df['Date'].apply(lambda d: d.to_pydatetime()) # Convert the date column to python datetime
df.info()
```

```python
df['Weekday'] = df['Date'].apply(lambda d: d.isoweekday()) # Extract the weekday from the date column
df.head()
```

```python
extract_year = lambda x: datetime.datetime.strptime(str(x), '%Y-%m-%d %H:%M:%S').year
```

```python
df['Year'] = df['Date'].apply(extract_year) # Extract the year from the date column
df.sample(10)
```

```python
extract_month = lambda x: datetime.datetime.strptime(str(x), '%Y-%m-%d %H:%M:%S').month
```

```python
df['Month'] = df['Date'].apply(extract_month) # Extract the month from the date column
df.sample(10)
```

```python
plt.figure(figsize=(12,8))
plt.plot(df['Date'], df['Close'], '-', color='blue')
plt.xlabel('Date')
plt.ylabel('AAPL Close Price')
plt.show();
```

```python
df_month = df.groupby('Month').mean() # Group the data by month and calculate the mean
df_month
```

```python
plt.figure(figsize=(12,8))
plt.plot(df_month.index, df_month['Close'], '-', color='blue')
plt.xlabel('Month')
plt.ylabel('AAPL Close Price')

plt.show();
```
